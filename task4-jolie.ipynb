```python
data["distance"] = pd.to_numeric(data["distance"], errors="coerce")
data["fare_amount"] = pd.to_numeric(data["fare_amount"], errors="coerce")
```

```python
data = data.dropna(subset=["distance", "fare_amount"])
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=data, x="distance", y="fare_amount", alpha=0.3)
plt.title("Fare vs Distance")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=data, x="Car Condition", y="fare_amount")
plt.title("Fare Amount by Car Condition")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=data, x="Weather", y="fare_amount")
plt.title("Fare Amount by Weather")
plt.show()
```

```python
num_cols = data.select_dtypes(include=['float64', 'int64']).columns
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data[num_cols].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
data = data.copy()
data["pickup_datetime"] = pd.to_datetime(data["pickup_datetime"], errors="coerce")
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=data, x="hour", y="fare_amount")
plt.title("Fare by Hour of Day")
plt.show()
```

```python
# Select categorical columns
cat_cols = data.select_dtypes(include=['object']).columns

# Only plot columns with reasonable unique values
unique_value_threshold = 50

for col in cat_cols:
    if data[col].nunique() <= unique_value_threshold:
        plt.figure(figsize=(10, 6))
        sns.countplot(data=data, x=col, order=data[col].value_counts().index)
        plt.title(f"Countplot of {col}")
        plt.xticks(rotation=45)
        plt.show()
    else:
        print(f"Skipping {col} (too many unique values: {data[col].nunique()})")
```
